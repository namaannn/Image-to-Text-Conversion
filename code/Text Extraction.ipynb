```python
import matplotlib.pyplot as plt
import re
from PIL import Image
import pytesseract
%matplotlib inline
```

```python
img=Image.open(r"C:\Users\naman\OneDrive\Pictures\test.JPG")
plt.imshow(img)
```

```python
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
```

```python
text_data=pytesseract.image_to_string(img.convert('RGB'),lang='eng')
```

```python
print(text_data)
```

```python
name=re.search("Name: (\w+)",text_data)
print(name[1])
```

```python
cord=re.search("Geo-Coordinates: (\S+)",text_data)
print(cord[1])
```
